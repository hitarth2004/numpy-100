```python
#List Operations
import statistics
import pandas as pd

# Specify the Excel file path
excel_file = ('/content/WQ1.xlsx')

# Read the Excel file into a DataFrame using pandas
df = pd.read_excel(excel_file)

# Convert the DataFrame column to a list
column_data = df['pH'].tolist()  # Replace 'ColumnName' with the actual column name

# Perform four operations on the list
operation1 = sum(column_data)
operation2 = max(column_data)
operation3 = min(column_data)
operation4 = statistics.mean(column_data)

# Print the results
print("Sum:", operation1)
print("Maximum:", operation2)
print("Minimum:", operation3)
print("Avg:", operation4)
```

```python
#Set Operations
import pandas as pd

# Read the Excel sheet into a pandas DataFrame
df = pd.read_excel('/content/WQ1.xlsx')

# Convert the DataFrame columns to sets
a = {}
for column in df.columns:
    a[column] = set(df[column])

# Print the sets
for column, column_set in a.items():
    print(f'{column}: {column_set}')

#1.For loop
for i in a:
  print(a)

#2.Clear
a.clear()
print("Clear Set:-",a)
```
